```python
# opens coordinate list

f=open('/Users/camfoltz/Desktop/GalacticScan.dat')
lines = f.readlines()
perp0data=[]
perp180data = []
indexarray =[]

# makes array of indexs of desired points

for i in range(5):
    perp0data.append(lines[i*73])
    indexarray.append(i*73)
for k in range(5):
    perp180data.append(lines[k*73 + 36])
    indexarray.append(k*73+36)
```

```python
# plots every data file for given indexes above

for index in indexarray:
    data = np.genfromtxt('/Users/camfoltz/Desktop/SRTFiles/DATA/GalacticScan1/Coordinate{}.dat'.format(index), dtype=float,
    delimiter="\t", autostrip=True,skip_header=1, names =True)
    x =[data[i][0] for i in range(data.size)]
    y =[data[i][1] for i in range (data.size)]
    plt.plot(x,y)
    #plots red line in center frequency
    plt.axvline(x=1.4204e9, color ='r')
    plt.ylim(-85,-78)
    plt.title(index)
    plt.show()
```
